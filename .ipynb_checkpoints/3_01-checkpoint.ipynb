```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def setup_grid(ax):
  ax.tick_params(axis='both', which='major', labelsize=100)
  ax.tick_params(axis='both', which='minor', labelsize=80)
  x_major_ticks = np.arange(0, 28 + 1 , 1)
  x_minor_ticks = np.arange(0, 28 + 0.1, 0.1)
  y_major_ticks = np.arange(0, 18 + 1, 1)
  y_minor_ticks = np.arange(0, 18 + 0.1 , 0.1)
  ax.xaxis.set_ticks(x_major_ticks)
  ax.yaxis.set_ticks(y_major_ticks)
  ax.xaxis.set_ticks(x_minor_ticks, minor=True)
  ax.yaxis.set_ticks(y_minor_ticks, minor=True)
  ax.grid(which='major', linewidth=5, c='grey')
  ax.grid(which='minor', linestyle='-', c='green')
  
```

```python
#Data without ring

line1_x = [2, 5, 5.4, 5.4, 4.4]
line2_x = [7, 9.5, 9.8, 9.2, 9.6]
line3_x = [14, 13.8, 13.5, 13.8, 13.5]
line4_x = [18.5, 18.5, 18.6, 18.5, 20]
line5_x = [24, 23.8, 23.3, 23.1, 23.5]

all_lines_y = np.arange(2, 18 + 4, 4)
df_x = pd.DataFrame(data = [line1_x, line2_x, line3_x, line4_x, line5_x],
                    index = ['3.40 V', '5.40 V', '7.40 V', '9.40 V', '11.40 V'])
df_y = pd.DataFrame(data = [all_lines_y, all_lines_y, all_lines_y, all_lines_y, all_lines_y],
                    index = ['3.40 V', '5.40 V', '7.40 V', '9.40 V', '11.40 V'])

```

```python
cm = 1/2.54
```

```python
fig = plt.figure(figsize=(28 * cm * 10, 18 * cm * 10))
ax = fig.add_subplot(1, 1, 1)

width=20



for i in range(len(df_x)):
  ax.plot(df_x.iloc[i], all_lines_y, linewidth=width)
  ax.scatter(df_x.iloc[i], all_lines_y)

#Scatter
ax.scatter(line1_x, all_lines_y, s=size)
ax.scatter(line2_x, all_lines_y, s=size)
ax.scatter(line3_x, all_lines_y, s=size)
ax.scatter(line4_x, all_lines_y, s=size)
ax.scatter(line5_x, all_lines_y, s=size)

setup_grid(ax)
plt.show()
```
